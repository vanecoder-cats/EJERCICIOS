# DETERMINAR SI UNA PALABRA ES PALÍNDROMO O NO

Ingresar palabra

Lenguaje Natural:
1. Colocar palabra
2. Leer palabra
3. Invertir palabra
4. Comparar la palabra invertida con la palabra original
5. mostrar si la palabra es palíndromo o no

# Diagrama de Flujo

Diagrama en blanco (1).jpeg



```python
palabra = input("Ingresar palabra: ")
palabra_invertida = palabra [::-1]

if palabra == palabra_invertida:
    print(f"La palabra es palíndromo")
else: print(f"La palabra no es palíndromo")
```
