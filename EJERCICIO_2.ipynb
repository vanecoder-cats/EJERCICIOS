# DEFINIR NÚMERO MAYOR
ENTRADA: Definir num1, num2 y num3

SALIDA: Mostrar número mayor
# Lenguaje Natural
1. Ingresar num1, num2 y num3
2. Leer num1, num2 y num3
3. Comparar los números
4. Determinar cual es mayor
5. Mostrar número mayor

# Diagrama de Flujo

Diagrama en blanco.jpeg

```python
num1 = input("Ingresar el primer número: ")
num2 = input("Ingresar el segundo número: ")
num3 = input("Ingresar el tercer número: ")

if num1 > num2 and num1 > num3:
  mayor = num1
elif num2 > num3 or num2 > num1:
  mayor = num2
else: mayor = num3

print(f"El número mayor es: {mayor}")
```
